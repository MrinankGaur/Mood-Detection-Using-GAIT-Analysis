```python
import pandas as pd
import joblib
from collections import Counter

# Load model + feature order
model = joblib.load("mood_model.pkl")
columns = joblib.load("feature_columns.pkl")
```

```python
input_data = [
    [90.0,0.722,0.1438610440668355,1090.8640625,957.8084375,3955.0,4095.0,2120817.7980834963,1857645.0392413088,1.007896100460484,1.8,1313,1260,2864.1359375,3137.1915625,5.172383333333334,2.343811666666667,4.595704456488569,6.28970033291443,0.2246316666666666,0.0386766666666666,1.4378819438085155,1.365948607057463,0.6025104577300818,1.1389167380988785]
]
```

```python
def predict_mood(input_data):
    
    # Convert to DataFrame
    df = pd.DataFrame(input_data, columns=columns)
    
    # Predict
    preds = model.predict(df)
    
    # Majority voting
    final_mood = Counter(preds).most_common(1)[0][0]
    
    return final_mood
```

```python
result = predict_mood(input_data)

print("Final Mood:", result)
```
